```python
import numpy as np

# Assuming you have loaded the Iris dataset into a DataFrame named 'iris'

# Define subsets based on species
iris_setosa = iris[iris['species'] == 'setosa']
iris_virginica = iris[iris['species'] == 'virginica']
iris_versicolor = iris[iris['species'] == 'versicolor']

# Calculate the standard deviation for petal_length for each class
std_setosa = np.std(iris_setosa["petal_length"])
std_virginica = np.std(iris_virginica["petal_length"])
std_versicolor = np.std(iris_versicolor["petal_length"])

print("Standard Deviation for Iris-setosa petal length:", std_setosa)
print("Standard Deviation for Iris-virginica petal length:", std_virginica)
print("Standard Deviation for Iris-versicolor petal length:", std_versicolor)
```

```python
import numpy as np

# Assuming you have loaded the Iris dataset into variables iris_setosa, iris_virginica, and iris_versicolor

# Calculate the means for petal_length for each class
mean_setosa = np.mean(np.append(iris_setosa["petal_length"], 50))
mean_virginica = np.mean(iris_virginica["petal_length"])
mean_versicolor = np.mean(iris_versicolor["petal_length"])

print("Mean for Iris-setosa petal length:", mean_setosa)
print("Mean for Iris-virginica petal length:", mean_virginica)
print("Mean for Iris-versicolor petal length:", mean_versicolor)
```
