```python
def tri_recursion(k):
    if(k>0):
        print('k value is ',k)
        print('Calling tthe dunction tri_recursion')
        fun_val=tri_recursion(k-1)
        print('value from the function ',fun_val)
        print('Adding ','k','fun_val')
        result =k+fun_val
        print(result)
    else:
        result=0
    return result

print("Recusrion Example Results")
tri_recursion(2)
```
