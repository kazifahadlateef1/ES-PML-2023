```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load training and test datasets
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')

# Extract target variable from the training set
y_train = df_train['Survived']

# Drop target variable and columns with non-numeric values
columns_to_drop = ['Survived', 'Name', 'Ticket', 'Cabin']
X_train = df_train.drop(columns_to_drop, axis=1)

# Combine training and test sets for consistent one-hot encoding
combined_data = pd.concat([X_train, df_test], axis=0, ignore_index=True)

# Identify categorical columns that need one-hot encoding
categorical_cols = ['Sex', 'Embarked']

# Perform one-hot encoding
encoder = OneHotEncoder(drop='first', sparse=False)
encoded_data = encoder.fit_transform(combined_data[categorical_cols])
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(categorical_cols))

# Concatenate the one-hot encoded features with the original data
combined_data = pd.concat([combined_data, encoded_df], axis=1)

# Drop the original categorical columns
combined_data.drop(categorical_cols, axis=1, inplace=True)

# Split the combined data back into training and test sets
X_train_encoded = combined_data.iloc[:len(X_train)]
X_test_encoded = combined_data.iloc[len(X_train):]

# Split the training data into training and validation sets
X_train_encoded, X_val_encoded, y_train, y_val = train_test_split(X_train_encoded, y_train, test_size=0.2, random_state=42)

def train_and_evaluate_knn(X_train, y_train, X_test, n_neighbors):
    # Initialization with the current value of n_neighbors
    neigh = KNeighborsClassifier(n_neighbors=n_neighbors)
    
    # Training
    neigh.fit(X_train, y_train)
    
    # Prediction
    Y_prediction_test = neigh.predict(X_test)
    
    return Y_prediction_test

# Iterate through different values of neighbors
for i in range(1, 25):
    Y_prediction_test = train_and_evaluate_knn(X_train_encoded, y_train, X_test_encoded, n_neighbors=i)
    
    # Example: Output the predictions for the test set using the current number of neighbors
    print(f'Predictions for test set with {i} neighbors:', Y_prediction_test)
```
