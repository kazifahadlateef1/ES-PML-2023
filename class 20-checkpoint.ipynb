```python
df_train['Age'].fillna(value=df_train['Age'].mean(),inplace=True)
df_train.isnull()
```

```python
df_train.describe()
```

```python
sns.histplot(x=df_train["Age"],label="Age",bins=10,kde=True)
plt.xlabel("S Age in years")
plt.ylabel("Density (count)")
plt.title("age distribution")

plt.legend()
plt.show()
```

```python
sns.histplot(x=df_train["Sex"],label="Gender",bins=10,kde=True)
plt.xlabel("Passange Gender")
plt.ylabel("Density (count)")
plt.title("Gender Distribution")

plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.FacetGrid(df_train, hue="Survived") \
    .map(sns.displot, "Parch") \
    .add_legend()

plt.show()
```

```python
sns.boxplot(x="Survived",y="Age",data=df_train)
plt.title("Box plot of age")
plt.xlabel("Survival status")
plt.ylabel("Age")
plt.show()
```

```python
sns.boxplot(x="Survived",y="Parch",data=df_train)
plt.title("Box plot of Parch")
plt.xlabel("Survival status")
plt.ylabel("Parch")
plt.show()
```

```python
sns.boxplot(x="Survived",y="Fare",data=df_train)
plt.title("Box plot of Fare")
plt.xlabel("Survival status")
plt.ylabel("Fare")
plt.show()
```

```python
sns.violinplot(x="Survived",y="Age",data=df_train)
plt.title("Violine plot of Age")
plt.show()
```

```python
sns.violinplot(x="Survived",y="Fare",data=df_train)
plt.title("Violine plot of Fare")
plt.show()
```
